```python
X_initial=df['X_pos'][0]
Y_initial=df['Y_pos'][0]
Xvel_initial=df['X_vel'][0]
Yvel_initial=df['Y_vel'][0]
```

```python
x=np.array([X_initial,Y_initial,Xvel_initial,Yvel_initial]).reshape(4,1)
```

```python
print("The state matrix",x) #The initial estimate state matrix
```

```python
#Kalman Filtering

#Step 1----- Updating the predicting state matrix: 
#->X_pred(n)=P*X(n-1) + Q*U(n)+W --- 
#P and Q are standard matrices to keep the order of the matrix same , X(n-1) is previous state matrix, U(n) is control variable matrix , W is predicted noise matrix
```

```python
df
```

```python
#Creating the Standard Matrix P
del_time=float(input("Enter the time interval between successive iterations:"))
P=np.zeros((4,4))
P[0,0],P[1,1],P[2,2],P[3,3]=[1,1,1,1]
P[0,2],P[1,3]=del_time,del_time
print(P)
#Standard Matrix P created



```

```python
#Creating the Standard Matrix Q
Q=np.zeros((4,2))
Q[0,0],Q[1,1]=(0.5)*(del_time)**2,(0.5)*(del_time)**2
Q[2,0],Q[3,1]=del_time,del_time
print(Q)
#Standard Matrix Q created
```

```python
#In our solution of the Kalman filter problem, we are considering an acceleration in the x and y direction: taken from user
acc_x=float(input("Enter the acceleration in the x-axis:"))
acc_y=float(input("Enter the acceleration in the y-axis:"))
U=np.zeros((2,1))
U[0,0],U[1,0]=acc_x,acc_y
print(U)
#Created the control Variable Matrix U
```

```python
def state_predmatrix(a_old,P,Q,U):
  a_prednew=np.dot(P,a_old)+np.dot(Q,U)
  return a_prednew
```

```python
#Step 2----- Updating the predicted covariance matrix: 
#->S_pred(n)= PS(n-1)P* --- 
#P is standard matrix to keep the order of the matrix same , S(n-1) is previous state covariance matrix
```

```python
sigma_X_pos=float(input("Enter the standard deviation of X_position:"))
sigma_Y_pos=float(input("Enter the standard deviation of Y_position:"))
sigma_X_vel=float(input("Enter the standard deviation of X_velocity:"))
sigma_Y_vel=float(input("Enter the standard deviation of Y_velocity:"))
S=np.zeros((4,4))

S[0,0],S[1,1],S[2,2],S[3,3]=(sigma_X_pos)**2,(sigma_Y_pos)**2,(sigma_X_vel)**2,(sigma_Y_vel)**2
```

```python
print(S)
```

```python
R=np.zeros((4,4))
R[0,0],R[1,1],R[2,2],R[3,3]=(sigma_X_pos)**2,(sigma_Y_pos)**2,(sigma_X_vel)**2,(sigma_X_vel)**2
```

```python
print(R)
```

```python
def cov_predmatrix(s_old,P):
  s_prednew=np.dot((np.dot(P,s_old)),np.transpose(P))
  return s_prednew
```

```python
#Step 4----- Calculating the Kalman Gain: 
#->K=(S_pred(n)*HT)/((H*S_pred(n)*HT)+R)
#H is standard matrix to keep the order of the matrix same , S_pred(n) is current predicted state covariance matrix
```

```python
H=np.zeros((4,4))
H[0,0],H[1,1],H[2,2],H[3,3] =(1,1,1,1)
HT=np.transpose(H)
# H
```

```python
def Kalmangain(s_prednew,H,R):
  K=np.dot(np.dot(s_prednew,HT),(np.linalg.inv(np.dot(np.dot(H,s_prednew),HT)+R)))
  return K
```

```python
def state_update(a_prednew,K,Y):
  a_new=a_prednew + np.dot(K,(Y-np.dot(H,a_prednew)))
  return a_new    
  
```

```python
D=np.zeros((4,4))
D[0,0],D[1,1],D[2,2],D[3,3]=1,1,1,1
print(D)
```

```python
def state_measurematrix(c_old,D):
  Y=np.dot(D,c_old)
  return Y
```

```python
def cov_update(pred_cov,K):
  I=np.eye(4)
  S=np.dot((I-np.dot(K,H)),pred_cov)
  return S
```
