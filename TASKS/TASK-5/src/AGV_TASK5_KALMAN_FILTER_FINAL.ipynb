```python
#Kalman Filtering

#Step 1----- Updating the predicting state matrix: 
#->X_pred(n)=P*X(n-1)+W --- 
#P and Q are standard matrices to keep the order of the matrix same , X(n-1) is previous state matrix, U(n) is control variable matrix , W is predicted noise matrix
```

```python
df
```

```python
#Creating the Standard Matrix P
del_time=float(input("Enter the time interval between successive iterations:"))
P=np.zeros((6,6))
P[0,0],P[1,1],P[2,2],P[3,3],P[4,4],P[5,5]=[1,1,1,1,1,1]
P[0,2],P[1,3],P[2,4],P[3,5]=del_time,del_time,del_time,del_time
P[0,4],P[1,5]=(0.5)*((del_time)**2),(0.5)*((del_time)**2)
print(P)
#Standard Matrix P created



```

```python
def state_predmatrix(a_old,P):
  a_prednew=np.dot(P,a_old)
  return a_prednew
```

```python
#Step 2----- Updating the predicted covariance matrix: 
#->S_pred(n)= PS(n-1)P* --- 
#P is standard matrix to keep the order of the matrix same , S(n-1) is previous state covariance matrix
```

```python
sigma_X_pos=float(input("Enter the standard deviation of X_position:"))
sigma_Y_pos=float(input("Enter the standard deviation of Y_position:"))
sigma_X_vel=float(input("Enter the standard deviation of X_velocity:"))
sigma_Y_vel=float(input("Enter the standard deviation of Y_velocity:"))
sigma_X_acc=float(input("Enter the standard deviation of X_acceleration:"))
sigma_Y_acc=float(input("Enter the standard deviation of Y_acceleration:"))
S=np.zeros((6,6))

S[0,0],S[1,1],S[2,2],S[3,3],S[4,4],S[5,5]=(sigma_X_pos)**2,(sigma_Y_pos)**2,(sigma_X_vel)**2,(sigma_Y_vel)**2,(sigma_X_acc)**2,(sigma_Y_acc)**2
```

```python
print(S)
```

```python
R=np.zeros((6,6))
R[0,0],R[1,1],R[2,2],R[3,3],R[4,4],R[5,5]=(sigma_X_pos)**2,(sigma_Y_pos)**2,(sigma_X_vel)**2,(sigma_Y_vel)**2,(sigma_X_acc)**2,(sigma_Y_acc)**2
```

```python
print(R)
```

```python
def cov_predmatrix(s_old,P):
  s_prednew=np.dot((np.dot(P,s_old)),np.transpose(P))
  return s_prednew
```

```python
#Step 4----- Calculating the Kalman Gain: 
#->K=(S_pred(n)*HT)/((H*S_pred(n)*HT)+R)
#H is standard matrix to keep the order of the matrix same , S_pred(n) is current predicted state covariance matrix
```

```python
H=np.zeros((6,6))
H[0,0],H[1,1],H[2,2],H[3,3],H[4,4],H[5,5] =(1,1,1,1,1,1)
HT=np.transpose(H)
# H
```

```python
def Kalmangain(s_prednew,H,R):
  K=np.dot(np.dot(s_prednew,HT),(np.linalg.inv(np.dot(np.dot(H,s_prednew),HT)+R)))
  return K
```

```python
def state_update(a_prednew,K,Y):
  a_new=a_prednew + np.dot(K,(Y-np.dot(H,a_prednew)))
  return a_new    
  
```

```python
D=np.zeros((6,6))
D[0,0],D[1,1],D[2,2],D[3,3],D[4,4],D[5,5]=1,1,1,1,1,1
print(D)
```

```python
def state_measurematrix(c_old,D):
  Y=np.dot(D,c_old)
  return Y
```

```python
def cov_update(pred_cov,K):
  I=np.eye(6)
  S=np.dot((I-np.dot(K,H)),pred_cov)
  return S
```

```python
state=state_predmatrix(x,P)
cov=cov_predmatrix(S,P)
l1=df['X_pos']
l2=df['Y_pos']
l3=df['X_vel']
l4=df['Y_vel']
count=0
time=0
t=[]
for i in range(1,360):
  state_final=np.zeros((1,6))
  count+=1
  time=time+del_time
  print("------------------------------------")
  print("The iteration count:",count)
  print("The time is:",time,"seconds")
  t.append(time)
  print(" ")
  A=np.array([l1[i],l2[i],l3[i],l4[i],state[4,0],state[5,0]]).reshape(6,1)
  y=state_measurematrix(A,D)
  K=Kalmangain(cov,H,R)
  state_final=state_update(state,K,y)
  print("The positions : ({},{})".format(state_final[0],state_final[1]))
  print(" ")
  print("The velocities : ({},{})".format(state_final[2],state_final[3]))
  print(" ")
  dfinal_s = dfinal_s.append(dict(zip(dfinal_s.columns,state_final)), ignore_index=True)
  state=state_final
  cov_final=cov_update(cov,K)
  print("The uncertainties of position: ({},{}".format(cov_final[0][0],cov_final[1][1]))
  print(" ")
  print("The uncertainties of velocities: ({},{}".format(cov_final[2][2],cov_final[3][3]))
  print(" ")
  cov=cov_final










```
